```python
# importing pandas library
import pandas as pd
```

```python
#creating pandas dataframe 
bikes=pd.read_csv('london_merged.csv')
```

```python
#look into data
bikes.info()
bikes.describe()
```

```python
bikes
```

```python
#changing humidity to be between 0 and 1
if bikes.hum.loc[1] > 1:
    bikes.hum = bikes.hum/100
bikes
```

```python
#dictionary to rename column names
col_rename = {'timestamp':'time',
              'cnt':'count',
              't1':'temp',
              't2':'temp_feel_like',
              'hum':'humidity'}
```

```python
#renaming the columns
bikes.rename(col_rename, axis =1, inplace=True )
bikes
```

```python
#changing season to descriptive values
#"season" - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.

season_map = {'0.0':'spring',
              '1.0':'summer',
              '2.0':'fall',
              '3.0':'winter'}

bikes.season = bikes.season.astype('str')

bikes.season = bikes.season.map(season_map)

bikes
```

```python
#Changind weather code to descriptive values
#"weathe_code" category description: 1 = Clear ; 2 = scattered clouds / few clouds 3 = Broken clouds 4 = Cloudy 7 = Rain/ light Rain shower/ Light rain 10 = rain with thunderstorm 26 = snowfall 94 = Freezing Fog

weather_map = {
    '1.0' : 'Clear',
    '2.0' : 'Scattered Clouds',
    '3.0': 'Broken Clouds',
    '4.0' : 'CLoudy',
    '7.0' : 'Rain',
    '10.0': 'Thunder',
    '26': 'Snowfall',
    '94' : 'Freezing Fog'
}


bikes.weather_code = bikes.weather_code.astype('str')

bikes.weather_code = bikes.weather_code.map(season_map)

bikes
```
